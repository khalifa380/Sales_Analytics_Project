```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("sample_superset.csv")

data.head()
```

```python
monthly_sales['Month_Number'] = range(len(monthly_sales))
```

```python
type(monthly_sales)
```

```python
monthly_sales = monthly_sales.reset_index()
```

```python
print(monthly_sales.head())
```

```python
print(monthly_sales.columns)
```

```python
monthly_sales['Month_Number'] = range(len(monthly_sales))
```

```python
monthly_sales.head()
```

```python
monthly_sales.head()
```

```python
print(monthly_sales.dtypes)
```

```python
monthly_sales['Sales'] = pd.to_numeric(monthly_sales['Sales'], errors='coerce')
```

```python
monthly_sales = monthly_sales.dropna()
```

```python
monthly_sales.head()
```

```python
from sklearn.linear_model import LinearRegression

X = monthly_sales[['Month_Number']]
y = monthly_sales['Sales']

model = LinearRegression()
model.fit(X, y)
```

```python
future_months = pd.DataFrame({
    'Month_Number': list(range(len(monthly_sales), len(monthly_sales)+6))
})

future_sales = model.predict(future_months)

print(future_sales)
```

```python
import matplotlib.pyplot as plt

plt.plot(monthly_sales['Month_Number'], monthly_sales['Sales'], label="Actual Sales")
plt.plot(future_months['Month_Number'], future_sales, linestyle="--", label="Forecast")

plt.title("Sales Forecast")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()

plt.show()
```

```python
category_sales = data.groupby("Category")["Sales"].sum().reset_index()

category_sales
```

```python
category_sales["Future_Sales"] = category_sales["Sales"] * 1.10
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(category_sales["Category"], category_sales["Future_Sales"])

plt.title("Future Demand by Category")
plt.xlabel("Category")
plt.ylabel("Predicted Future Sales")

plt.show()
```

```python
top_category = category_sales.loc[category_sales["Future_Sales"].idxmax()]

print("Category with highest future demand:", top_category["Category"])
```

```python
product_sales = data.groupby("Product Name")["Sales"].sum().reset_index()

top_products = product_sales.sort_values(by="Sales", ascending=False).head(10)

plt.barh(top_products["Product Name"], top_products["Sales"])

plt.title("Top Products Likely to Drive Future Sales")

plt.show()
```
